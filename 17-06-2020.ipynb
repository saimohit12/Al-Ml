# OneHotEncoder-->similar to pd.get_dummies

```python
---> Import OneHotEncoder Class
--->Creates an object for the class
--->Column Transformer-->insider this columntransform,we will pass the OneHotEncoder
--->it is two mandatory arguments-->transforms,remainder
--->transform argument,it will accept the list of tuple values
```

```python
## Importiong libraries
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
#Inside the columntransformer,we have to pass the OneHotEncoder
column_transformer=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough')
```

```python
##objectname.fit_transform(data)
```

```python
data=column_transformer.fit_transform(data)
data
```

```python
#creating dataframe out of array
x=pd.DataFrame(data)
x
```
